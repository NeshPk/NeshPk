#####  【Problem 1 】 Random number creation

```python
import numpy as np

# Set the seed for reproducibility
np.random.seed(0)

# Define the mean and covariance matrix
mean = np.array([-3, 0])
covariance = np.array([[1.0, 0.8], [0.8, 1.0]])

# Generate 500 random numbers from the multivariate normal distribution
random_numbers = np.random.multivariate_normal(mean, covariance, 500)

# Print shape of the generated array
print(random_numbers.shape)
```

##### 【Problem 2 】 Visualization by spray diagram

```python
import matplotlib.pyplot as plt

# Extract x and y coordinates for the scatter plot
x = random_numbers[:, 0]
y = random_numbers[:, 1]

# Create the scatter plot
plt.scatter(x, y)

# Add labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot of Random Numbers")

# Show the plot
plt.show()
```

##### 【Problem 3 】 Visualization by histogram

```python
# Extract x and y coordinates
x = random_numbers[:, 0]
y = random_numbers[:, 1]

# Create histograms for each dimension
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(x, bins=20)
plt.xlabel("X")
plt.ylabel("Frequency")
plt.title("Histogram of X")
plt.xlim(min(min(x),min(y)), max(max(x),max(y)))

plt.subplot(1, 2, 2)
plt.hist(y, bins=20)
plt.xlabel("Y")
plt.ylabel("Frequency")
plt.title("Histogram of Y")
plt.xlim(min(min(x),min(y)), max(max(x),max(y)))

plt.tight_layout()
plt.show()
```

#####【Problem 4 】 Add data

```python
# Set the seed for reproducibility
np.random.seed(0)

# Define the mean and covariance matrix for Problem 1
mean1 = np.array([-3, 0])
covariance = np.array([[1.0, 0.8], [0.8, 1.0]])

# Generate 500 random numbers for Problem 1
random_numbers1 = np.random.multivariate_normal(mean1, covariance, 500)

# Define the mean and covariance matrix for Problem 4
mean4 = np.array([0, -3])

# Generate 500 random numbers for Problem 4
random_numbers4 = np.random.multivariate_normal(mean4, covariance, 500)

# Extract x and y coordinates for Problem 1
x1 = random_numbers1[:, 0]
y1 = random_numbers1[:, 1]

# Extract x and y coordinates for Problem 4
x4 = random_numbers4[:, 0]
y4 = random_numbers4[:, 1]

# Create the scatter plot
plt.scatter(x1, y1, label="Problem 1")
plt.scatter(x4, y4, label="Problem 4")

# Add labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot of Random Numbers")

# Add legend
plt.legend()

# Show the plot
plt.show()
```

##### 【Problem 5 】 Data coupling

```python

# Concatenate the two arrays
combined_data = np.concatenate((random_numbers1, random_numbers4), axis=0)

# Print the shape of the combined array
print(combined_data.shape)
```

##### 【Problem 6 】 Labeling

```python

combined_data = np.concatenate((random_numbers1, random_numbers4), axis=0)

# Create labels for each data point (0 for Problem 1, 1 for Problem 4)
labels = np.array([0] * 500 + [1] * 500)

# Combine data and labels into a single array
combined_data_with_labels = np.column_stack((combined_data, labels))

# Print the shape of the combined array with labels
print(combined_data_with_labels.shape)
```
