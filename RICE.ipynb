##### 【Problem 1 】 Number of rice grains on day 100

```python
import matplotlib.pyplot as plt


#Calculates the number of rice grains received each day and the cumulative total.
#Args:days (int): The number of days to calculate.
#Returns: A tuple containing lists of daily rice grains and cumulative rice grains.

def rice_grain_calculator(days):

  daily_rice = []
  accumulated_rice =  []
  rice_grain = 1 #start from 1
  sum_grains = 0 #start from 0

  for day in range (days):
    daily_rice.append (rice_grain)
    sum_grains += rice_grain
    accumulated_rice.append (sum_grains)
    rice_grain *= 2 ## Double the rice grains for the following day

  return daily_rice, accumulated_rice

#Number of rice grains on day 100

days = 100
daily_rice, accumulated_rice = rice_grain_calculator(days)

print(f"Total rice gained day {days}: {accumulated_rice[-1]}")

# Plotting Rice gain over time
plt.figure(figsize=(8, 5))
plt.plot(range(1, days + 1), daily_rice,)
plt.plot(range(1, days + 1), accumulated_rice, color='r')
plt.xlabel('Day')
plt.ylabel('Rice Grain')
plt.title('Rice Grain Growth Over Time')
plt.legend(['Daily Rice', 'Accumulated Rice'])
plt.grid(True)
plt.show()

# from around day  85 Both the "Daily Rice Grains" and "Cumulative Rice Grains" lines demonstrate a very rapid, exponential increase.
```

#####【Problem 2 】 Number of rice grains other than 100 days


```python
def compute_sorori_shinzaemon(n_days=100):

    list_n_grains = [] #list of the number of rice grains received each day (length equals n_days)
    list_total_grains = [] #list of the cumulative number of rice grains received by a given day
    current_grain = 1
    total_grains = 0

    for day in range(1, n_days + 1): #Number of days receiving rice
        list_n_grains.append(current_grain)
        total_grains += current_grain
        list_total_grains.append(total_grains)
        current_grain *= 2

    return list_n_grains, list_total_grains

list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=10)

#A function that displays a graph of the number of rice grains per day and the cumulative number of rice grains.
def plot_rice_grains(days, daily_rice, accumulated_rice ):

  plt.figure(figsize=(8, 5))
  plt.plot(range(1, days + 1), daily_rice,)
  plt.plot(range(1, days + 1), accumulated_rice, color='r')
  plt.xlabel('Day')
  plt.ylabel('Rice Grain')
  plt.title('Rice Grain Growth Over Time')
  plt.legend(['Daily Rice', 'Accumulated Rice'])
  plt.grid(True)
  plt.show()

#In the case of 30 days
days_30 = 30
daily_grains_30, cumulative_grains_30 = compute_sorori_shinzaemon(days_30)
print(f"Total rice gained day {days_30}: {cumulative_grains_30[-1]}")
plot_rice_grains(days_30, daily_grains_30, cumulative_grains_30)

#In the case of 80 days
days_80 = 80
daily_grains_80, cumulative_grains_80 = compute_sorori_shinzaemon(days_80)
print(f"Total rice gained day {days_80}: {cumulative_grains_80[-1]}")
plot_rice_grains(days_80, daily_grains_80, cumulative_grains_80)

```

#####【Problem 3 】 How many people can live on rice they can have

```python
#Calculates how many days a group of people can live on a given amount of rice.
#total_rice_grains (int): The total number of rice grains available.
#num_people (int): The number of people in the group.
#return The number of days the group can live on the rice.

def calculate_days_of_sustenance(total_rice_grains, num_people):
  grains_per_person_per_day = 40000 #Estimated rice grains needed per person per day:
  total_grains_per_day = grains_per_person_per_day * num_people #Total rice grains needed per day for the group:

  ## Calculate the number of days the rice will last:
  if total_grains_per_day == 0:
    return float('inf')
  days_of_sustenance = total_rice_grains // total_grains_per_day
  return days_of_sustenance

#Example on a case scenario
total_grains = 1000000000 #Define total_grains here
people = 10 #Define number of people here


days = calculate_days_of_sustenance(total_grains, people)
print(f"{people} people can live for {days} days with {total_grains:,} rice grains.")
```
