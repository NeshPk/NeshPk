#####【Problem 】 Chestnut bun covers the solar system

```python
#Return int: The time (in minutes) it takes to cover the target volume.
#Return list: A list of volumes at each doubling interval.
#Return list: A list of times at each doubling interval.

import matplotlib.pyplot as plt

def time_to_cover_distance(initial_volume, desired_volume, time_to_double=5):
  volume_now = initial_volume
  time_passed = 0
  volume_levels = [volume_now]
  times = [time_passed]

  if initial_volume >= desired_volume:
    return 0, volume_levels, times

  while volume_now < desired_volume:
    volume_now *= 2
    time_passed += time_to_double
    volume_levels.append(volume_now)
    times.append(time_passed)

  return time_passed, volume_levels, times

#list Volumes
chestnut_bun_volume = 1e-5
solar_system_volume = 4.5e40
tokyo_dome_volume = 1240000
ball_volume = 0.00418879

#Time for solar system
time_solar, volumes_solar, times_solar = time_to_cover_distance(chestnut_bun_volume, solar_system_volume)
print(f"Time to cover the solar system: {time_solar} minutes")

# Time for Tokyo Dome
time_tokyo, volumes_tokyo, times_tokyo = time_to_cover_distance(chestnut_bun_volume, tokyo_dome_volume)
print(f"Time to fill the Tokyo Dome: {time_tokyo} minutes")

# Calculate for a single ball doubling.
time_ball, volumes_ball, times_ball = time_to_cover_distance(ball_volume, tokyo_dome_volume)
print(f"Time to fill the Tokyo Dome with doubling balls: {time_ball} minutes")


```

```python
#Ploting Solar System Plot
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1) #creates the individual subplots.
plt.plot(times_solar, volumes_solar)
plt.yscale('log')
plt.title("Chestnut Bun vs. Solar System")
plt.xlabel("Time (minutes)")
plt.ylabel("Volume (cubic meters)")
```

```python
# Tokyo Dome Plot
plt.subplot(1, 3, 2)
plt.plot(times_tokyo, volumes_tokyo)
plt.yscale('log')
plt.title("Chestnut Bun vs. Tokyo Dome")
plt.xlabel("Time (minutes)")
plt.ylabel("Volume (cubic meters)")
```

```python
# Tokyo Dome with Ball Plot
plt.subplot(1, 3, 3)
plt.plot(times_ball, volumes_ball)
plt.yscale('log')
plt.title("Ball vs. Tokyo Dome")
plt.xlabel("Time (minutes)")
plt.ylabel("Volume (cubic meters)")

plt.tight_layout()
plt.show()
```
