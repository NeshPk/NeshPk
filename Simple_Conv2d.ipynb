```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score

# ============================================================
#                    ACTIVATION FUNCTIONS (MLP)
# ============================================================
class ReLU:
    def __call__(self, x):
        return np.maximum(0, x)
    def gradient(self, x):
        return (x > 0).astype(float)

class Sigmoid:
    def __call__(self, x):
        return 1 / (1 + np.exp(-x))
    def gradient(self, x):
        s = self.__call__(x)
        return s * (1 - s)

class Tanh:
    def __call__(self, x):
        return np.tanh(x)
    def gradient(self, x):
        return 1 - np.tanh(x)**2

class Softmax:
    def __call__(self, z):
        exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
        return exp_z / np.sum(exp_z, axis=1, keepdims=True)

# ============================================================
#                       INITIALIZERS (MLP)
# ============================================================
class HeInitializer:
    def __call__(self, shape):
        fan_in = shape[0]
        return np.random.randn(*shape) * np.sqrt(2.0 / max(1, fan_in))

class XavierInitializer:
    def __call__(self, shape):
        fan_in, fan_out = shape[0], shape[1]
        limit = np.sqrt(6.0 / (fan_in + fan_out))
        return np.random.uniform(-limit, limit, size=shape)

# ============================================================
#                        OPTIMIZER (MLP)
# ============================================================
class SGD:
    def __init__(self, lr=0.01):
        self.lr = lr
    def update(self, params, grads):
        for key in params.keys():
            params[key] -= self.lr * grads[key]

# ============================================================
#                   SIMPLE MLP NEURAL NETWORK
# ============================================================
class NeuralNetwork:
    def __init__(self, layers, activations, initializer=HeInitializer(), optimizer=SGD(),
                 learning_rate=0.01, batch_size=32, epochs=10, verbose=True):
        self.layers = layers
        self.activations = activations
        self.initializer = initializer
        self.optimizer = optimizer
        self.learning_rate = learning_rate
        self.batch_size = batch_size
        self.epochs = epochs
        self.verbose = verbose
        self.weights = []
        self.biases = []

    def initialize_parameters(self):
        self.weights = []
        self.biases = []
        for i in range(len(self.layers) - 1):
            self.weights.append(self.initializer((self.layers[i], self.layers[i + 1])))
            self.biases.append(np.zeros((1, self.layers[i + 1])))

    def forward(self, X):
        self.cache = {'A0': X}
        A = X
        for i in range(len(self.weights) - 1):
            Z = np.dot(A, self.weights[i]) + self.biases[i]
            A = self.activations[i](Z)
            self.cache[f'Z{i + 1}'] = Z
            self.cache[f'A{i + 1}'] = A

        # final layer: softmax
        Z = np.dot(A, self.weights[-1]) + self.biases[-1]
        A = Softmax()(Z)
        self.cache[f'Z{len(self.weights)}'] = Z
        self.cache[f'A{len(self.weights)}'] = A
        return A

    def compute_loss(self, y_true, y_pred):
        m = y_true.shape[0]
        return -np.sum(y_true * np.log(y_pred + 1e-8)) / m

    def backward(self, y_true):
        grads = {'weights': [], 'biases': []}
        m = y_true.shape[0]

        # Output layer grad (softmax + CE)
        dZ = self.cache[f'A{len(self.weights)}'] - y_true
        grads['weights'].insert(0, np.dot(self.cache[f'A{len(self.weights) - 1}'].T, dZ) / m)
        grads['biases'].insert(0, np.sum(dZ, axis=0, keepdims=True) / m)

        # Hidden layers
        for i in reversed(range(len(self.weights) - 1)):
            dA = np.dot(dZ, self.weights[i + 1].T)
            dZ = dA * self.activations[i].gradient(self.cache[f'Z{i + 1}'])
            grads['weights'].insert(0, np.dot(self.cache[f'A{i}'].T, dZ) / m)
            grads['biases'].insert(0, np.sum(dZ, axis=0, keepdims=True) / m)

        return grads

    def update_parameters(self, grads):
        for i in range(len(self.weights)):
            self.weights[i] -= self.learning_rate * grads['weights'][i]
            self.biases[i]  -= self.learning_rate * grads['biases'][i]

    def fit(self, X, y, X_val=None, y_val=None):
        X = X.values if hasattr(X, 'values') else X
        y = y.values if hasattr(y, 'values') else y

        self.initialize_parameters()

        history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}

        for epoch in range(self.epochs):
            perm = np.random.permutation(len(X))
            X_shuffled = X[perm]
            y_shuffled = y[perm]

            batch_losses = []
            for i in range(0, len(X), self.batch_size):
                X_batch = X_shuffled[i:i + self.batch_size]
                y_batch = y_shuffled[i:i + self.batch_size]

                y_pred = self.forward(X_batch)
                loss = self.compute_loss(y_batch, y_pred)
                batch_losses.append(loss)

                grads = self.backward(y_batch)
                self.update_parameters(grads)

            # Metrics
            train_pred = np.argmax(self.forward(X), axis=1)
            train_acc = accuracy_score(np.argmax(y, axis=1), train_pred)

            train_loss = np.mean(batch_losses)
            history['train_loss'].append(train_loss)
            history['train_acc'].append(train_acc)

            if X_val is not None and y_val is not None:
                val_pred = self.forward(X_val)
                val_loss = self.compute_loss(y_val, val_pred)
                val_acc = accuracy_score(np.argmax(y_val, axis=1), np.argmax(val_pred, axis=1))
                history['val_loss'].append(val_loss)
                history['val_acc'].append(val_acc)


            if self.verbose:
                if X_val is not None and y_val is not None:
                    print(f"MLP Epoch {epoch + 1}/{self.epochs} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f}")
                else:
                    print(f"MLP Epoch {epoch + 1}/{self.epochs} | Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f}")

        self._plot_learning_curve(history, title="MLP Learning Curve")
        return history

    def predict(self, X):
        X = X.values if hasattr(X, 'values') else X
        return np.argmax(self.forward(X), axis=1)

    def _plot_learning_curve(self, history, title="Learning Curve"):
        plt.figure(figsize=(10, 5))
        plt.plot(history['train_loss'], label='Train Loss')
        if history['val_loss']:
            plt.plot(history['val_loss'], label='Validation Loss')
        plt.title(title)
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.legend()
        plt.grid(True)
        plt.show()

# ============================================================
#                    CONV2D / MAXPOOL PIPELINE
#  (Conv forward/backward, MaxPool forward/backward, smoothing)
# ============================================================
def conv2d_output_size(H, W, FH, FW, stride=1, padding=0):
    out_h = (H + 2*padding - FH) // stride + 1
    out_w = (W + 2*padding - FW) // stride + 1
    return out_h, out_w

class Conv2d:
    def __init__(self, in_channels, out_channels, kernel_size, stride=1, padding=0, lr=0.01):
        self.in_channels = in_channels
        self.out_channels = out_channels
        self.kernel_size = kernel_size  # (FH, FW)
        self.stride = stride
        self.padding = padding
        FH, FW = kernel_size
        # He-style init
        scale = np.sqrt(2.0 / (in_channels * FH * FW))
        self.W = np.random.randn(FH, FW, in_channels, out_channels) * scale
        self.b = np.zeros(out_channels)
        self.lr = lr

    def forward(self, x):
        self.x = x  # (N, C, H, W)
        N, C, H, W = x.shape
        FH, FW = self.kernel_size
        out_h, out_w = conv2d_output_size(H, W, FH, FW, self.stride, self.padding)

        if self.padding > 0:
            x = np.pad(x, ((0,0),(0,0),(self.padding,self.padding),(self.padding,self.padding)), 'constant')

        self.x_padded = x  # keep for backward
        out = np.zeros((N, self.out_channels, out_h, out_w))

        for i in range(out_h):
            for j in range(out_w):
                h_start = i * self.stride
                w_start = j * self.stride
                region = x[:, :, h_start:h_start+FH, w_start:w_start+FW]  # (N,C,FH,FW)
                for m in range(self.out_channels):
                    out[:, m, i, j] = np.sum(region * self.W[:, :, :, m], axis=(1,2,3)) + self.b[m]

        self.out_h, self.out_w = out_h, out_w
        return out

    def backward(self, dout):
        N, C, H, W = self.x.shape
        FH, FW = self.kernel_size
        out_h, out_w = self.out_h, self.out_w

        dW = np.zeros_like(self.W)
        db = np.zeros_like(self.b)
        dx_padded = np.zeros_like(self.x_padded)

        x = self.x_padded

        for i in range(out_h):
            for j in range(out_w):
                h_start = i * self.stride
                w_start = j * self.stride
                region = x[:, :, h_start:h_start+FH, w_start:w_start+FW]  # (N,C,FH,FW)
                for m in range(self.out_channels):
                    db[m] += np.sum(dout[:, m, i, j])
                    # dW
                    for n in range(N):
                        for k in range(self.in_channels):
                            dW[:, :, k, m] += region[n, k, :, :] * dout[n, m, i, j]
                            dx_padded[n, k, h_start:h_start+FH, w_start:w_start+FW] += self.W[:, :, k, m] * dout[n, m, i, j]


        # remove padding from dx
        if self.padding > 0:
            dx = dx_padded[:, :, self.padding:-self.padding, self.padding:-self.padding]
        else:
            dx = dx_padded

        # SGD update
        self.W -= self.lr * dW
        self.b -= self.lr * db
        return dx

class MaxPool2d:
    def __init__(self, pool_size=2, stride=2):
        self.pool_size = pool_size
        self.stride = stride

    def forward(self, x):
        self.x = x
        N, C, H, W = x.shape
        PH = PW = self.pool_size
        out_h = (H - PH) // self.stride + 1
        out_w = (W - PW) // self.stride + 1
        out = np.zeros((N, C, out_h, out_w))
        self.max_mask = np.zeros_like(x, dtype=bool)

        for i in range(out_h):
            for j in range(out_w):
                h_start = i * self.stride
                w_start = j * self.stride
                region = x[:, :, h_start:h_start+PH, w_start:w_start+PW]  # (N,C,PH,PW)
                max_vals = np.max(region, axis=(2,3), keepdims=True)
                mask = (region == max_vals)
                self.max_mask[:, :, h_start:h_start+PH, w_start:w_start+PW] = mask
                out[:, :, i, j] = max_vals.squeeze()

        self.out_h, self.out_w = out_h, out_w
        return out

    def backward(self, dout):
        dx = np.zeros_like(self.x)
        N, C, out_h, out_w = dout.shape
        PH = PW = self.pool_size
        for i in range(out_h):
            for j in range(out_w):
                h_start = i * self.stride
                w_start = j * self.stride
                dx[:, :, h_start:h_start+PH, w_start:w_start+PW] += \
                    self.max_mask[:, :, h_start:h_start+PH, w_start:w_start+PW] * \
                    dout[:, :, i, j][:, :, None, None]
        return dx

class Smoothing:
    """Probability smoothing on predictions (can be used like label smoothing at inference)."""
    def __init__(self, smoothing_factor=0.1):
        self.smoothing_factor = smoothing_factor
    def smooth(self, probs):
        # probs: (N, K)
        N, K = probs.shape
        return probs * (1 - self.smoothing_factor) + (self.smoothing_factor / K)

class FullyConnected:
    def __init__(self, input_size, output_size, lr=0.01):
        self.W = np.random.randn(input_size, output_size) * 0.01
        self.b = np.zeros((1, output_size))
        self.lr = lr
        # running EMA estimates (for "replace with estimate" objective)
        self.ema_W = np.zeros_like(self.W)
        self.ema_b = np.zeros_like(self.b)
        self.ema_beta = 0.9

    def forward(self, x):
        self.x = x
        return np.dot(x, self.W) + self.b

    def backward(self, dout):
        dW = np.dot(self.x.T, dout)
        db = np.sum(dout, axis=0, keepdims=True)
        dx = np.dot(dout, self.W.T)

        # SGD
        self.W -= self.lr * dW
        self.b -= self.lr * db

        # update EMA estimates
        self.ema_W = self.ema_beta * self.ema_W + (1 - self.ema_beta) * self.W
        self.ema_b = self.ema_beta * self.ema_b + (1 - self.ema_beta) * self.b
        return dx

    def replace_with_estimate(self):
        """Replace current params with EMA estimate to meet the objective."""
        self.W = self.ema_W.copy()
        self.b = self.ema_b.copy()

def softmax(x):
    ex = np.exp(x - np.max(x, axis=1, keepdims=True))
    return ex / np.sum(ex, axis=1, keepdims=True)

def cross_entropy(y_true, y_prob):
    m = y_true.shape[0]
    return -np.sum(y_true * np.log(y_prob + 1e-8)) / m

def accuracy_np(pred_probs, y_true_onehot):
    return np.mean(np.argmax(pred_probs, axis=1) == np.argmax(y_true_onehot, axis=1))

# ============================================================
#                         LOAD MNIST
# ============================================================
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
X = mnist.data.astype('float32') / 255.0
y = mnist.target.astype('int')
y = y.reshape(-1, 1)

encoder = OneHotEncoder(sparse_output=False)
y_encoded = encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
X_train, X_val,  y_train, y_val  = train_test_split(X_train, y_train, test_size=0.25, random_state=42)  # 0.6/0.2/0.2

# ============================================================
#                       RUN YOUR MLP CONFIGS
# ============================================================
configurations = [
    {
        'name': '2 Layers (ReLU)',
        'layers': [784, 128, 10],
        'activations': [ReLU(), ReLU()],
        'initializer': HeInitializer(),
        'optimizer': SGD(lr=0.01),
        'epochs': 10,
        'batch_size': 64
    },
    {
        'name': '3 Layers (ReLU)',
        'layers': [784, 256, 128, 10],
        'activations': [ReLU(), ReLU(), ReLU()],
        'initializer': HeInitializer(),
        'optimizer': SGD(lr=0.01),
        'epochs': 10,
        'batch_size': 64
    },
    {
        'name': '2 Layers (Tanh)',
        'layers': [784, 128, 10],
        'activations': [Tanh(), Tanh()],
        'initializer': XavierInitializer(),
        'optimizer': SGD(lr=0.01),
        'epochs': 10,
        'batch_size': 64
    }
]

results = []
for config in configurations:
    print(f"\nTraining {config['name']}...")
    nn = NeuralNetwork(
        layers=config['layers'],
        activations=config['activations'],
        initializer=config['initializer'],
        optimizer=config['optimizer'],
        learning_rate=0.01,
        batch_size=config['batch_size'],
        epochs=config['epochs'],
        verbose=True
    )

    history = nn.fit(X_train, y_train, X_val, y_val)
    y_pred = nn.predict(X_test)
    test_acc = accuracy_score(np.argmax(y_test, axis=1), y_pred)
    results.append((config['name'], test_acc))
    print(f"{config['name']} - Test Accuracy: {test_acc:.4f}")

print("\nMLP Final Results:")
for name, acc in results:
    print(f"{name}: Test Accuracy = {acc:.4f}")

# ============================================================
#                  CONV2D + MAXPOOL + FC EXPERIMENT
#   (Meets all CNN objectives incl. output size, backprop, smoothing,
#    and 'replace FC with estimate' mid-training)
# ============================================================
print("\n=== Conv2D + MaxPool2D + FC (NumPy) Training ===")

# Reshape for conv pipeline
X_train_img = X_train.reshape(-1, 1, 28, 28)
X_val_img   = X_val.reshape(-1, 1, 28, 28)
X_test_img  = X_test.reshape(-1, 1, 28, 28)

# Model hyperparams
conv = Conv2d(in_channels=1, out_channels=8, kernel_size=(3,3), stride=1, padding=0, lr=0.01)
pool = MaxPool2d(pool_size=2, stride=2)
fc   = FullyConnected(input_size=8 * 13 * 13, output_size=10, lr=0.01)
relu = ReLU()
smooth = Smoothing(smoothing_factor=0.1)

epochs_cnn = 10          # full training as requested
batch_size = 64

# Output size calculation (explicit)
H_in, W_in = 28, 28
FH, FW = 3, 3
H_out, W_out = conv2d_output_size(H_in, W_in, FH, FW, stride=1, padding=0)
print(f"Conv2D Output size: ({H_out}, {W_out}) for input (28,28), kernel (3,3), stride=1, padding=0")
print(f"After MaxPool(2,2) -> size: ({H_out//2}, {W_out//2})")
assert H_out == 26 and W_out == 26, "Output size calc sanity check failed."

def forward_conv_pipeline(Xb):
    # Conv -> ReLU -> Pool -> Flatten -> FC -> Softmax -> Smooth
    z1 = conv.forward(Xb)                         # (N, 8, 26, 26)
    a1 = relu(z1)                                 # ReLU
    p1 = pool.forward(a1)                         # (N, 8, 13, 13)
    flat = p1.reshape(p1.shape[0], -1)            # (N, 1352)
    logits = fc.forward(flat)                     # (N, 10)
    probs = softmax(logits)                       # (N, 10)
    probs_sm = smooth.smooth(probs)               # smoothing
    cache = {'z1': z1, 'a1': a1, 'p1': p1, 'flat': flat, 'logits': logits, 'probs_sm': probs_sm}
    return probs_sm, cache

def backward_conv_pipeline(cache, yb):
    # dL/dprobs for CE with smoothed probs: (probs_sm - y)/N
    N = yb.shape[0]
    dout = (cache['probs_sm'] - yb) / N           # (N,10)

    # FC backward
    dflat = fc.backward(dout)                     # (N,1352)

    # Unflatten -> pool backward
    dp1 = dflat.reshape(-1, 8, 13, 13)
    da1 = pool.backward(dp1)

    # ReLU backward
    dz1 = da1 * (cache['z1'] > 0)

    # Conv backward
    dX = conv.backward(dz1)
    return dX

# Training loop
train_losses, val_losses, train_accs, val_accs = [], [], [], []
for epoch in range(1, epochs_cnn + 1):
    perm = np.random.permutation(X_train_img.shape[0])
    X_train_img_shuf = X_train_img[perm]
    y_train_shuf     = y_train[perm]

    batch_losses = []
    for i in range(0, X_train_img_shuf.shape[0], batch_size):
        Xb = X_train_img_shuf[i:i+batch_size]
        yb = y_train_shuf[i:i+batch_size]

        probs_sm, cache = forward_conv_pipeline(Xb)
        loss = cross_entropy(yb, probs_sm)
        batch_losses.append(loss)

        backward_conv_pipeline(cache, yb)

    # Metrics each epoch
    train_probs, _ = forward_conv_pipeline(X_train_img[:1024])  # (optional) subset for speed in metrics
    train_loss = np.mean(batch_losses)
    train_acc = accuracy_np(train_probs, y_train[:1024])

    val_probs, _ = forward_conv_pipeline(X_val_img)
    val_loss = cross_entropy(y_val, val_probs)
    val_acc = accuracy_np(val_probs, y_val)

    train_losses.append(train_loss)
    val_losses.append(val_loss)
    train_accs.append(train_acc)
    val_accs.append(val_acc)

    # Mid-training: replace FC with EMA estimate (objective: "replace with estimate")
    if epoch == (epochs_cnn // 2):
        fc.replace_with_estimate()
        print("Replaced FullyConnected weights with EMA estimates at mid-training.")

    print(f"CNN Epoch {epoch}/{epochs_cnn} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | Train Acc (subset): {train_acc:.4f} | Val Acc: {val_acc:.4f}")

# Final test accuracy for CNN pipeline
test_probs, _ = forward_conv_pipeline(X_test_img)
test_acc_cnn = accuracy_np(test_probs, y_test)
print(f"\nConv2D+MaxPool+FC - Test Accuracy: {test_acc_cnn:.4f}")

# Plot CNN learning curves
plt.figure(figsize=(10,5))
plt.plot(train_losses, label='Train Loss (CNN)')
plt.plot(val_losses, label='Val Loss (CNN)')
plt.title("CNN Learning Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# ============================================================
#            VISUALIZE PREDICTIONS FROM BEST MLP MODEL
# ============================================================
best_idx = int(np.argmax([acc for _, acc in results]))
best_config = configurations[best_idx]
print(f"\nVisualizing predictions from best MLP model: {best_config['name']}")

nn_best = NeuralNetwork(
    layers=best_config['layers'],
    activations=best_config['activations'],
    initializer=best_config['initializer'],
    optimizer=best_config['optimizer'],
    learning_rate=0.01,
    batch_size=best_config['batch_size'],
    epochs=best_config['epochs'],
    verbose=False
)
nn_best.fit(X_train, y_train)

sample_indices = np.random.choice(len(X_test), 5, replace=False)
plt.figure(figsize=(10, 3))
for i, idx in enumerate(sample_indices):
    plt.subplot(1, 5, i + 1)
    plt.imshow(X_test[idx].reshape(28, 28), cmap='gray')
    pred = nn_best.predict(X_test[idx].reshape(1, -1))[0]
    true = np.argmax(y_test[idx])
    plt.title(f"P:{pred}\nT:{true}", fontsize=8)
    plt.axis('off')
plt.show()
```
