##### 【Problem 1 】 Primary function

```python
import numpy as np

# Problem 1: Primary function
x = np.arange(-50, 50.1, 0.1)
y = 1/2 * x**2 + 1

print("x shape:", x.shape)
print("y shape:", y.shape)
```

##### 【Problem 2 】 Array coupling

```python
# Problem 2: Array coupling
xy_combined = np.column_stack((x, y))

print("Combined array shape:", xy_combined.shape)
print("First 5 rows of combined array:\n", xy_combined[:5])
```

##### 【Problem 3 】 Request slope

```python
# Problem 3: Calculate slope
x_diff = np.diff(x)
y_diff = np.diff(y)
slope = y_diff / x_diff

print("Slope shape:", slope.shape)
print("First 5 slopes:", slope[:5])
```

##### 【Problem 4 】 Draw a graph

```python
# Problem 4: Draw graphs
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))

# Graph of the primary function
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.plot(x, y)
plt.title("Primary Function: y = 1/2 * x^2 + 1")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

# Graph of the slope
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.plot(x[:-1], slope)  # x[:-1] to match the length of slope
plt.title("Slope (Gradient)")
plt.xlabel("x")
plt.ylabel("Slope")
plt.grid(True)

plt.tight_layout()  # Adjust subplot parameters to give specified padding.
plt.show()
```

##### 【Problem 5 】 Python functionalization

```python
import numpy as np

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):

    x = np.arange(x_range[0], x_range[1], x_range[2])
    y = function(x)
    array_xy = np.column_stack((x, y))
    gradient = np.diff(y) / np.diff(x)
    return array_xy, gradient

def function1(array_x):
    array_y = array_x**2
    return array_y

array_xy, gradient = compute_gradient(function1)

print("array_xy shape:", array_xy.shape)
print("gradient shape:", gradient.shape)
print("gradient first 5 values:", gradient[:5])

def function2(array_x):
  array_y = 2*array_x**2+2*array_x
  return array_y

array_xy, gradient = compute_gradient(function2)

print("array_xy shape:", array_xy.shape)
print("gradient shape:", gradient.shape)
print("gradient first 5 values:", gradient[:5])

def function3(array_x):
  array_y = np.sin(array_x**0.5)
  return array_y

array_xy, gradient = compute_gradient(function3, x_range=(0,50.1,0.1))

print("array_xy shape:", array_xy.shape)
print("gradient shape:", gradient.shape)
print("gradient first 5 values:", gradient[:5])
```

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):

    x = np.arange(x_range[0], x_range[1], x_range[2])
    y = function(x)
    array_xy = np.column_stack((x, y))
    gradient = np.diff(y) / np.diff(x)
    return array_xy, gradient

def function1(array_x):
    return array_x**2

def function2(array_x):
    return 2 * array_x**2 + 2 * array_x

def function3(array_x):
    return np.sin(array_x**0.5)


array_xy1, gradient1 = compute_gradient(function1)
print("Function 1 Gradient (first 5):", gradient1[:5])

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(array_xy1[:, 0], array_xy1[:, 1])
plt.title("Function: x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(array_xy1[:-1, 0], gradient1)
plt.title("Gradient")
plt.xlabel("x")
plt.ylabel("Slope")
plt.grid(True)

plt.tight_layout()
plt.show()


array_xy2, gradient2 = compute_gradient(function2)
print("Function 2 Gradient (first 5):", gradient2[:5])


plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(array_xy2[:, 0], array_xy2[:, 1])
plt.title("Function: 2x^2 + 2x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(array_xy2[:-1, 0], gradient2)
plt.title("Gradient")
plt.xlabel("x")
plt.ylabel("Slope")
plt.grid(True)

plt.tight_layout()
plt.show()

array_xy3, gradient3 = compute_gradient(function3, x_range=(0, 50.1, 0.1))
print("Function 3 Gradient (first 5):", gradient3[:5])

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(array_xy3[:, 0], array_xy3[:, 1])
plt.title("Function: sin(x^0.5)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(array_xy3[:-1, 0], gradient3)
plt.title("Gradient")
plt.xlabel("x")
plt.ylabel("Slope")
plt.grid(True)

plt.tight_layout()
plt.show()
```

##### 【Problem 6 】 Request minimum value

```python

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    x = np.arange(x_range[0], x_range[1], x_range[2])
    y = function(x)
    array_xy = np.column_stack((x, y))
    gradient = np.diff(y) / np.diff(x)
    return array_xy, gradient

def function1(array_x):
    return array_x**2

def function2(array_x):
    return 2 * array_x**2 + 2 * array_x

def function3(array_x):
    return np.sin(array_x**0.5)

def find_min_and_display(function, x_range=(-50, 50.1, 0.1)):
    array_xy, gradient = compute_gradient(function, x_range)
    y_values = array_xy[:, 1]
    min_y = np.min(y_values)
    min_index = np.argmin(y_values)
    x_values = array_xy[:,0]

    print(f"Minimum y value: {min_y}")
    print(f"Minimum y index: {min_index}")
    print(f"x at min y: {x_values[min_index]}")

    if min_index > 0:
        print(f"Slope before min y: {gradient[min_index - 1]}")
    else:
        print("No slope before min y (at start of range).")

    if min_index < len(gradient):
        print(f"Slope after min y: {gradient[min_index]}")
    else:
        print("No slope after min y (at end of range).")

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(array_xy[:, 0], array_xy[:, 1])
    plt.title("Function")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.scatter(array_xy[min_index,0], array_xy[min_index,1], color='red') #Mark min point

    plt.subplot(1, 2, 2)
    plt.plot(array_xy[:-1, 0], gradient)
    plt.title("Gradient")
    plt.xlabel("x")
    plt.ylabel("Slope")
    plt.grid(True)
    plt.scatter(array_xy[min_index,0], gradient[min_index], color='red') #Mark slope at min

    plt.tight_layout()
    plt.show()

# Function 1
print("Function 1 (x^2):")
find_min_and_display(function1)

# Function 2
print("\nFunction 2 (2x^2 + 2x):")
find_min_and_display(function2)

# Function 3
print("\nFunction 3 (sin(x^0.5)):")
find_min_and_display(function3, x_range=(0, 50.1, 0.1))
```
